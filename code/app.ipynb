```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import json
import re
import os
import time
LOCATION_TEXT = input("Enter the location you want to search: ")
SEARCH_LIST_URL = "https://geocode.arcgis.com/arcgis/rest/services/World/GeocodeServer/findAddressCandidates?singleLine="+LOCATION_TEXT+"&f=json&outSR=%7B%22wkid%22%3A102100%7D"
```

```python
response = requests.get(SEARCH_LIST_URL)
with open('data.json', 'w') as outfile:
    json.dump(response.json(), outfile)
print(response.json())
print("Done")
```
